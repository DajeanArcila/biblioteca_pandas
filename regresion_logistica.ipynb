```python
#REGRESION LOGISTICA BASICA

import seaborn as sns
from sklearn.linear_model import LogisticRegression

import pandas as pd
```

```python
df= pd.read_excel('/content/golf_df.xlsx')
```

```python
df.head()
```

```python
#normalizamos las variables explicativas de nuestra target
```

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()

le.fit(df.clima)
df.clima= le.transform(df.clima)

le.fit(df.temp)
df.temp= le.transform(df.temp)

le.fit(df.vento)
df.vento= le.transform(df.vento)

le.fit(df.umidade)
df.umidade= le.transform(df.umidade)

le.fit(df.jogar)
df.jogar= le.transform(df.jogar)
```

```python
x = df.iloc[:,:-1]
y = df[['jogar']]
#AQUI DEFINIMOS NUESTRA VARIABLE TARGET
```

```python
model = LogisticRegression()
model.fit(x,y)
```

```python
predict= model.predict(x)
```

```python
x['predict']= predict

y= pd.DataFrame(y)

x['real']= y['jogar']
#AQUI COMPARAMOS LO REAL COM LO PRONOSTICADO(OSEA LO QUE HIZO EL MODELO)
x.head()
```
