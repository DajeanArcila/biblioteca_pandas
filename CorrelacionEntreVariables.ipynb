```python
import pandas as pd
import numpy as np

# Generar datos aleatorios para una base de datos con 50 filas y algunas variables
np.random.seed(42)

data = {
    'Ventas': np.random.randint(100, 1000, 50),
    'Precio': np.random.uniform(10, 100, 50),
    'Publicidad': np.random.randint(1000, 10000, 50),
    'Descuento': np.random.uniform(5, 25, 50),
    'Satisfacción_Cliente': np.random.uniform(1, 5, 50)
}

# Crear DataFrame
df = pd.DataFrame(data)



df.head()

```

```python
df.head(50)
```

```python
# Calcular la correlación entre 'ventas' y las otras variables
correlacion = df.corr()

```

```python
# Mostrar la correlación entre todas las variables (incluyendo ventas)
print(correlacion)
```

```python

# Si quieres ver solo la correlación de 'ventas' con las otras
print(correlacion['Ventas'])
```
