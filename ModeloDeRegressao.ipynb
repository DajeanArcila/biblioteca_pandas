#Comparando investimentos e iniciando o modelo de Regresao linear

```python
import matplotlib.pyplot as plt
```

```python
categorias = ['youtube', 'facebook', 'newspaper']
sales = [355, 59, 121]  # investimendo en cada lugar

plt.bar(categorias, sales, color=['blue', 'green', 'red'])

plt.xlabel('Categorías')
plt.ylabel('sales')
plt.title('Comparacao de vendas por categoría')

plt.show()
```

```python
from sklearn.linear_model import LinearRegression

# definir as variables de entrada(investimentos) e saida(sales)
investimento = dataset[['youtube', 'facebook', 'newspaper']]
vendas =dataset ['sales']

# Crear el modelo de regresión lineal
modelo = LinearRegression()

# Ajustar o modelo aos dados do investimento e vendas
modelo.fit(investimento, vendas)

# os coeficientes da regresao (retorno do investimento)
print("Coeficientes de regressão (retorno do investimento):")
print("youTube:", modelo.coef_[0])
print("facebook:", modelo.coef_[1])
print("newspaper:", modelo.coef_[2])
```

```python
sns.regplot(x=dataset["youtube"], y=dataset["sales"], label="youtube")
sns.regplot(x=dataset["facebook"], y=dataset["sales"], label="facebook")
sns.regplot(x=dataset["newspaper"], y=dataset["sales"], label="newspaper")


plt.xlabel('categorias')
plt.ylabel('sales')
plt.title('relacao entre sales e categorias')

plt.legend()

plt.show()
```

```python
# Calcular o coeficiente de determinacao R^2
coeficiente_r2 = modelo.score(dataset[['youtube', 'facebook', 'newspaper']], dataset['sales'])
print("Coeficiente de determinación R^2:", coeficiente_r2)

# predicoes de vendas usando o modelo
predicciones_ventas = modelo.predict(nuevas_inversiones[['youtube', 'facebook', 'newspaper']])
print("Predicciones de ventas para nuevas inversiones:")
print(predicciones_ventas)
```
