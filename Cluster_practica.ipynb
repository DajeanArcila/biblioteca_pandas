```python
import pandas as pd
import numpy as np

# Configuración de la semilla para reproducibilidad
np.random.seed(42)

# Número de empleados
num_employees = 1000

# Generar datos ficticios
data = {
    'Edad': np.random.randint(22, 60, size=num_employees),
    'Salario': np.random.randint(30000, 120000, size=num_employees),
    'Satisfacción Laboral': np.random.randint(1, 11, size=num_employees),
    'Horas Trabajadas': np.random.randint(30, 60, size=num_employees),
    'Departamento': np.random.choice(['Ventas', 'Finanzas', 'Marketing',], size=num_employees),
    'Antigüedad': np.random.randint(1, 20, size=num_employees)
}

# Crear DataFrame
df_employees = pd.DataFrame(data)

# Mostrar las primeras filas del DataFrame
print(df_employees.head())

# Guardar en un archivo CSV
df_employees.to_csv('datos_empleados.csv', index=False)
```

```python
df = df_employees
df.head()
```

```python
df.info()
```

```python
df = pd.get_dummies(df, columns=['Departamento'], drop_first=False)
```

```python
# https://matplotlib.org/stable/tutorials/introductory/customizing.html
sns.set_theme(
    context='talk',
    style='ticks',
    font_scale=.8,
    palette='tab10',
    rc={
        'figure.figsize': (12,8),
        'axes.grid': True,
        'grid.alpha': .2,
        'axes.titlesize': 'x-large',
        'axes.titleweight': 'bold',
        'axes.titlepad': 20,
    }
)

scatter_kwargs = dict(palette='viridis', alpha=0.8, linewidth=0)
```

```python
X= df[['Salario','Edad']]
y= df['Satisfacción Laboral']
```

```python
# Calcular el Within-Cluster Sum of Squares (WCSS) para diferentes valores de k
wcss = []
for k in range(1, 11):
    kmeans = KMeans(n_clusters=k, init='k-means++', max_iter=300, n_init=10, random_state=42)
    kmeans.fit(X)
    wcss.append(kmeans.inertia_)

# Graficar el método del codo
plt.figure(figsize=(10,6))
plt.plot(range(1, 11), wcss, marker='o', linestyle='--')
plt.title('Método del Codo')
plt.xlabel('Número de Clústeres')
plt.ylabel('WCSS')
plt.show()
```

```python

# Seleccionar columnas relevantes
X = df[['Edad','Salario','Satisfacción Laboral','Horas Trabajadas','Antigüedad']]
# Preprocesar los datos
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# Determinar el número de clusters (por ejemplo, 3)
kmeans = KMeans(n_clusters=4, random_state=0)
df['cluster'] = kmeans.fit_predict(X_scaled)

# Visualizar los resultados (usando 2D o 3D si es necesario)
plt.scatter(df['Salario'], df['Edad'], c=df['cluster'], cmap='viridis')
plt.xlabel('Salario')
plt.ylabel('Edad')
plt.title('Clusters')
plt.show()
```
