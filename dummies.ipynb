```python
import pandas as pd
import seaborn as sns
import numpy as np

```

```python
limite = pd.read_excel('/content/C├│pia de Limite_Credito_Dummy.xlsx')
```

```python
limite.head()
```

```python
limite = pd.get_dummies(data=limite, columns=['Escolaridade', 'Gênero',"Região"], prefix = ['Escolaridade', 'Região',"Gênero"])
```

```python
limite.head()
```

```python
limite.corr()
```

```python
import statsmodels.formula.api as smf
```

```python
function = 'LimitedoChequeEspecial~LimitedeCreditoImediato+Salario+RendimentoTotal+Idade+Escolaridade_Fundamental_Medio+Escolaridade_Superior_Pos+Região_RJ+Região_SP+Gênero_F+Gênero_M'
model= smf.ols(formula= function, data=limite).fit()
print(model.summary())
```

```python
#modelo MRLS
function = 'LimitedoChequeEspecial~RendimentoTotal+Salario+LimitedeCreditoImediato+Escolaridade_Fundamental_Medio+Escolaridade_Superior_Pos+Região_RJ+Região_SP+Gênero_F+Gênero_M'
model = smf.ols(formula=function, data=limite).fit()
print(model.summary())
```
