```python
import xmltodict
import json

# Nombre del archivo XML
xml_file = '/content/43240636552901000133550010000008311000011689-procNFe.xml'
# Nombre del archivo JSON de salida
json_file = 'archivo.json'

# Abrir y leer el archivo XML
with open(xml_file, 'r') as f:
    xml_data = f.read()

# Convertir XML a diccionario
xml_dict = xmltodict.parse(xml_data)

# Convertir diccionario a JSON
json_data = json.dumps(xml_dict, indent=4)

# Escribir JSON en archivo
with open(json_file, 'w') as f:
    f.write(json_data)

print(f"Archivo JSON guardado como {json_file}")

```
